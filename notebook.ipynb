## Task 4

## Time to close by Location.

When we look at time to close by location, we can see that all locations have a large range of outlying claims which take much longer to close than the average. Aside from outliers, each location has a similar interquartile range and similar values for both the upper and lower whiskers. This implies that the majority of claims get processed and closed in a similar amount of time across all four locations.

![Screen Shot 2023-02-15 at 4.56.20 PM](Screen%20Shot%202023-02-15%20at%204.56.20%20PM.png)

Based upon the above analysis, my first recommendation is that any intervention to improve time taken to close claims would be most impactful at *Recife* as this location has far more claims than any other. Recife has more claims that Forteleza and Natal combined; increasing performance at this location would affect the most number of customers. Recife has a similar performance to the other locations but if it performed *better* than the others, a larger percentage of customers would have a shorter wait time for claims to close. Any additional expenditures to increase performance (hiring more staff, or better technology for example) would have the largest return on investment at the Recife location.

My second recomendation would be to conduct a further investigation into the outlying claims that take longer than 273 days to close. Although these represent a minority of cases, they represent customers who have had to wait double (nearly tripple in the most extreme cases) the amount of time that other customers have had to wait. A benefit of reducing the number of outliers is that it would have a strong impact on the mean value of time to close across all claims. For example, if the outliers in the current dataset were to be removed (all claims longer than 273 days), the new mean would be 145 days which is a 22% decrease from 186 days. Mean values are common metrics used to assess performance and by limiting any future incidences of outliers it would ensure that this key metric does not get inflated negatively. I recommend conducting a review of all current claims that are approaching the 204 day mark (the 75th percentile) and ensuring that, if possible, these are closed in a timely manner.
